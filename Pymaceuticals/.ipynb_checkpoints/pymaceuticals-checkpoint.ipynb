```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data_complete = pd.merge(mouse_drug_data, clinical_trial_data, how="left", on=["Mouse ID"])


# Display the data table for preview
data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_tumor_volume_data = data_complete.groupby(["Drug", "Timepoint"])
mean_tumor_volume_data = grouped_tumor_volume_data["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
summary_table_01 = pd.DataFrame({"Tumor Volume (mm3)": mean_tumor_volume_data})

# Preview DataFrame
summary_table_01_complete = summary_table_01.reset_index()
summary_table_01_complete
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint 
grouped_tumor_volume_data = data_complete.groupby(["Drug", "Timepoint"])
sem_tumor_volume_data = grouped_tumor_volume_data["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
summary_table_02 = pd.DataFrame({"Tumor Volume (mm3)": sem_tumor_volume_data})

# Preview DataFrame
summary_table_02_complete = summary_table_02.reset_index()
summary_table_02_complete
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_pivot = summary_table_01_complete.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

sem_pivot = summary_table_02_complete.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

# Preview that Reformatting worked
# mean_pivot
sem_pivot
```

```python
# Generate the Plot (with Error Bars)

x_axis = np.arange(0, 50, 10)
x_limit = 45

plt.errorbar(mean_pivot.index, mean_pivot["Capomulin"], yerr=sem_pivot["Capomulin"],\
             color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(mean_pivot.index, mean_pivot["Infubinol"], yerr=sem_pivot["Infubinol"],\
             color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(mean_pivot.index, mean_pivot["Ketapril"],yerr=sem_pivot["Ketapril"],\
             color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(mean_pivot.index, mean_pivot["Placebo"],yerr=sem_pivot["Placebo"],\
             color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

plt.grid(linestyle="dashed")

plt.grid(linestyle="dashed")
plt.legend(loc='best', fontsize=12, fancybox=True)

# Save the Figure
plt.savefig("Output1.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Metastatic Site Data Grouped by Drug and Timepoint 
grouped_tumor_volume_data = data_complete.groupby(["Drug", "Timepoint"])
mean_met_site_data = grouped_tumor_volume_data["Metastatic Sites"].mean()

# Convert to DataFrame
summary_table_03 = pd.DataFrame({"Metastatic Sites": mean_met_site_data})

# Preview DataFrame
summary_table_03_complete = summary_table_03.reset_index()
summary_table_03_complete
```

```python
# Store the Standard Error associated with Metastatic Sites Grouped by Drug and Timepoint 
grouped_tumor_volume_data = data_complete.groupby(["Drug", "Timepoint"])
sem_met_site_data = grouped_tumor_volume_data["Metastatic Sites"].sem()

# Convert to DataFrame
summary_table_04 = pd.DataFrame({"Metastatic Sites": sem_met_site_data})

# Preview DataFrame
summary_table_04_complete = summary_table_04.reset_index()
summary_table_04_complete
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_pivot_met_site = summary_table_03_complete.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

sem_pivot_met_site = summary_table_04_complete.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
# mean_pivot_met_site
sem_pivot_met_site
```

```python
# Generate the Plot (with Error Bars)

x_axis = np.arange(0, 50, 10)
x_limit = 45

plt.errorbar(mean_pivot_met_site.index, mean_pivot_met_site["Capomulin"], yerr=sem_pivot_met_site["Capomulin"],\
             color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(mean_pivot_met_site.index, mean_pivot_met_site["Infubinol"], yerr=sem_pivot_met_site["Infubinol"],\
             color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(mean_pivot_met_site.index, mean_pivot_met_site["Ketapril"],yerr=sem_pivot_met_site["Ketapril"],\
             color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(mean_pivot_met_site.index, mean_pivot_met_site["Placebo"],yerr=sem_pivot_met_site["Placebo"],\
             color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")

plt.grid(linestyle="dashed")

plt.grid(linestyle="dashed")
plt.legend(loc='best', fontsize=12, fancybox=True)

# Save the Figure
plt.savefig("Output2.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric) 
grouped_tumor_volume_data = data_complete.groupby(["Drug", "Timepoint"])
mouse_count_data = grouped_tumor_volume_data["Mouse ID"].count()

# Convert to DataFrame
summary_table_05 = pd.DataFrame({"Mouse Count": mouse_count_data})

# Preview DataFrame
summary_table_05_complete = summary_table_05.reset_index()
summary_table_05_complete
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mouse_pivot = summary_table_05_complete.pivot(index="Timepoint", columns="Drug", values="Mouse Count")

# Preview that Reformatting worked
count_mouse_pivot
```

```python
# Generate the Plot (Accounting for percentages)

x_axis = np.arange(0, 50, 10)
x_limit = 45

plt.plot(count_mouse_pivot.index, (count_mouse_pivot["Capomulin"] * 4),\
             color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(count_mouse_pivot.index, (count_mouse_pivot["Infubinol"] * 4),\
             color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(count_mouse_pivot.index, (count_mouse_pivot["Ketapril"] * 4),\
             color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(count_mouse_pivot.index, (count_mouse_pivot["Placebo"] * 4),\
             color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

plt.grid(linestyle="dashed")

plt.grid(linestyle="dashed")
plt.legend(loc='best', fontsize=12, fancybox=True)

# Save the Figure
plt.savefig("Output3.png")

# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
drug_pct_change_mean = ((mean_pivot.iloc[-1] - mean_pivot.iloc[0]) / mean_pivot.iloc[0]) * 100

# Display the data to confirm
drug_pct_change_mean
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_change = (drug_pct_change_mean["Capomulin"], 
               drug_pct_change_mean["Infubinol"], 
               drug_pct_change_mean["Ketapril"], 
               drug_pct_change_mean["Placebo"])

x_axis = np.arange(0,len(pct_change),1)

width = 0.5

# Splice the data between passing and failing drugs

plt.bar(x_axis[0] - width, drug_pct_change_mean["Capomulin"],\
             color='g', align="center")
plt.bar(x_axis[1] - width, drug_pct_change_mean["Infubinol"],\
             color='r', align="center")
plt.bar(x_axis[2] - width, drug_pct_change_mean["Ketapril"],\
             color='r', align="center")
plt.bar(x_axis[3] - width, drug_pct_change_mean["Placebo"],\
             color='r', align="center")

# Orient widths. Add labels, tick marks, etc. 

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Capomulin", "Infubinol", "Ketapril", "Placebo"])

plt.xlim(-1.00, len(x_axis)+0.25)

plt.ylim(-25, 65)

plt.grid(True)

# Use functions to label the percentages of changes


# Call functions to implement the function calls


# Save the Figure
plt.savefig("Output4.png")

# Show the Figure
plt.show()
```
